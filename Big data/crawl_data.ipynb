# Crawl Data for many years

```python
# Thêm thư mục gốc dự án vào sys.path
import sys
import os
from pathlib import Path

# Xác định thư mục gốc dự án
notebook_dir = Path(os.getcwd())
project_root = notebook_dir.parent  # Đi lên 1 cấp từ scripts
sys.path.append(str(project_root))

print(f"Added {project_root} to Python path")

# Import các module
from data.collector import F1Collector
import config

# Tạo instance của F1Collector
collector = F1Collector(cache_enabled=True)

# Danh sách các năm cần thu thập dữ liệu
years = range(2019, 2026)  # 2019 đến 2025

# Thu thập dữ liệu cho từng năm
for year in years:
    print(f"\n{'='*50}")
    print(f"Collecting data for {year} season...")
    print(f"{'='*50}")
    
    try:
        # Thu thập dữ liệu cho toàn bộ mùa giải
        results = collector.collect_season_data(year)
        
        # Hiển thị tổng số Grand Prix đã thu thập
        num_gps = len(results)
        print(f"\nCollected data for {num_gps} Grand Prix in {year}")
        
        # Hiển thị tên các Grand Prix đã thu thập
        print(f"Grand Prix collected: {', '.join(results.keys())}")
    except Exception as e:
        print(f"Error collecting data for {year}: {str(e)}")
        continue

print("\nData collection completed!")

# Kiểm tra kết quả
print("\nFiles created:")
for year in years:
    year_dir = config.RAW_DATA_DIR / str(year)
    if year_dir.exists():
        files = list(year_dir.glob("*.parquet"))
        print(f"\nYear {year}: {len(files)} files")
```

# Kiểm tra số file đã tạo và tốc độ đọc file của Spark

```python
from pyspark.sql import SparkSession
import time
import os
from pathlib import Path

# Khởi tạo Spark session
spark = SparkSession.builder \
    .appName("F1 Parquet Read Test") \
    .config("spark.sql.legacy.parquet.nanosAsLong", "true") \
    .getOrCreate()

# Đường dẫn thư mục data/raw
raw_data_dir = Path(r"D:\VSCODE\UEH_BigData_Final\UEH_BigData_Final\data\raw")

# Các loại file cần đọc
file_types = ["_r_laps", "_r_weather", "_r_results", "_r_drivers"]

# Lấy danh sách các năm trong thư mục data/raw
years = [d.name for d in raw_data_dir.iterdir() if d.is_dir() and d.name.isdigit()]

# Kết quả lưu trữ
results = []

for year in years:
    year_path = raw_data_dir / year
    for file_type in file_types:
        # Tìm tất cả file parquet theo pattern
        files = list(year_path.glob(f"*{file_type}*.parquet"))
        if not files:
            continue
        
        # Chuyển danh sách Path thành danh sách string
        file_paths = [str(f) for f in files]
        
        # Đo thời gian đọc
        start_time = time.time()
        df = spark.read.parquet(*file_paths)
        
        # Thực hiện action để đảm bảo dữ liệu được đọc
        count = df.count()
        end_time = time.time()
        
        # Tính kích thước dữ liệu
        size_bytes = sum(f.stat().st_size for f in files)
        size_mb = size_bytes / (1024 * 1024)
        
        # Lưu kết quả
        results.append({
            "year": year,
            "file_type": file_type,
            "num_files": len(files),
            "num_rows": count,
            "size_mb": round(size_mb, 2),
            "read_time_sec": round(end_time - start_time, 2),
            "read_speed_mbps": round(size_mb / (end_time - start_time), 2) if (end_time - start_time) > 0 else "N/A"
        })

# In kết quả dạng bảng
print(f"{'Year':<6} {'Type':<12} {'Files':<6} {'Rows':<8} {'Size(MB)':<10} {'Time(s)':<10} {'Speed(MB/s)':<12}")
print("-" * 70)
for r in results:
    print(f"{r['year']:<6} {r['file_type']:<12} {r['num_files']:<6} {r['num_rows']:<8} {r['size_mb']:<10.2f} {r['read_time_sec']:<10.2f} {r['read_speed_mbps']}")

# Tính tổng
total_size_mb = sum(r['size_mb'] for r in results)
total_time_sec = sum(r['read_time_sec'] for r in results)
total_rows = sum(r['num_rows'] for r in results)
total_files = sum(r['num_files'] for r in results)

print("-" * 70)
print(f"TOTAL: {total_files} files, {total_rows} rows, {total_size_mb:.2f} MB, {total_time_sec:.2f} seconds")
print(f"Average read speed: {total_size_mb/total_time_sec:.2f} MB/s")

# Đóng Spark session
spark.stop()
```
